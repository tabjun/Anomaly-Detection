```python
result_df = pd.DataFrame()

# 각 모델별로 df_copy['f'heart_rate{i}']번째 값을 train
for j in range(1,1001):
    model = ExponentialSmoothing(
        df_copy[f'heart_rate{j}'], trend='additive', seasonal='additive', seasonal_periods=365).fit()

    # i번째 validation셋의 결과를 result{i}에 저장
    result_i = model.predict(
        start=df_copy.index[0], end=df_copy.index[-1])

    # result{i}를 result_df에 누적
    result_df[f'pred_y{j}'] = result_i
    # if j == 1:
    #     result_df[f'pred_y{j}'] = result_i
    # else:
    #     result_df[f'pred_y{j}'] = pd.concat([result_df, result_i], axis=1)
        
result_df.head()
```

```python
aa1 = result_df.copy()
aa2 = df_copy.copy()

aa1 = aa1.reset_index()
aa2 = aa2.reset_index()

diff_df = pd.merge(aa1, aa2, left_index=True, right_index=True,
                   how='inner')
# diff_df = aa2.merge(aa1, how='inner', on='index')
# diff_df['diff'] = diff_df['heart_rate1'] - diff_df['is_anomaly1']
print(diff_df)
```

```python
PERIOD = 12
GAMMA = 0
SF = 1.96
UB = []           
LB = []

for j in range(0,1000):
    # 실측치와 예측치를 비교하는 자료구조
    difference_array = []
    dt = []
    difference_table = {"actual": df_copy[f'heart_rate{j+1}'], "predicted": result_df[f'pred_y{j+1}'], "difference": difference_array, "UB": UB, "LB": LB}

# brutlag 알고리즘
# i안에 넣어버리면 j가 1일 동안에 i 1~1000, j가 2일 때 반복 한번에 넣어야함
for j,i in zip(range(0,1000),range(len(result_df))):
    # 밑에 식으로 하면 서로 다른 데이터프레임의 열로 연산하는거라 Nan만 출력함
    # 같은 데이터프레임으로 합쳐줘야함
    # diff = df_copy.iloc[f'heart_rate{j+1}']-result_df.iloc[:,i]
    diff_df[f'diff{j+1}'] = diff_df[f'heart_rate{j+1}'] - diff_df[f'pred_y{j+1}']
    if i < PERIOD:
        dt.append(GAMMA*abs(diff_df[f'diff{j+1}']))
    else:
        dt.append(GAMMA*abs(diff_df[f'diff{j+1}']) + (1-GAMMA)*dt[i-PERIOD])

    difference_array.append(diff_df[f'diff{j+1}'])
    UB.append(diff_df[f'pred_y{j+1}']+SF*dt[i])
    LB.append(diff_df[f'pred_y{j+1}']-SF*dt[i])
    
if (diff_df[f'diff{j+1}'] > UB[j]).any() or (diff_df[f'diff{j+1}'] < LB[j]).any():
    result_df[f'event{j+1}'] = 1
else:
    result_df[f'event{j+1}'] = 0
```
