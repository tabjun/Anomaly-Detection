```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima_model import ARIMA
# Load the dataframe
df = pd.read_csv('../data/heart_rate_tj.csv')

# Create a copy of the dataframe
df_copy = df[['update_time','heart_rate']].copy()

# Initialize variables
best_aic = float("inf")
best_order = None

# Iterate over different values of p and q
for i in range(1, 6):
    for j in range(1, 6):
        try:
            model = ARIMA(df_copy['heart_rate'], order=(i, 0, j))
            results = model.fit()
            if results.aic < best_aic:
                best_aic = results.aic
                best_order = (i, 0, j)
        except:
            continue
        
# Print the best order and AIC value
print(f'Best order (p, d, q): {best_order} with AIC: {best_aic}')
```

```python
# 반복문으로 p,q별 aic값 확인
for i in range(1, 6):
    for j in range(1, 6):
        try:
            model = ARIMA(df_copy['heart_rate'], order=(i, 0, j))
            results = model.fit()
            print(f'Order: {(i, 0, j)} with AIC: {results.aic}')
        except:
            continue
```
