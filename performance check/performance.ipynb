```python
import pandas as pd
import numpy as np
import math
```

```python
def performance(TP, TN, FP, FN):
    
    N = TP+TN+FP+FN
    
    Sensitivity=(TP/(TP+FN))*100

    Specificity=(1-FP/(TN+FP))*100

    Accuracy=((TP+TN)/(TP+TN+FP+FN))*100

    PPV= (TP/(TP+FP))*100 

    NPV= (TN/(TN+FN))*100 
    
    result = {
        'Sensitivity': [Sensitivity],
        'Specificity': [Specificity],
        'Accuracy':[Accuracy],
        'PPV':[PPV],
        'NPV':[NPV]
    }
    
    result_2 = pd.DataFrame(result)
    
    return result_2
```

```python
ewma_result = pd.DataFrame()

for i in range(1,1001):

    TP = len(df[(df[f'event{i}']==1)&(df[f'pred_y{i}']==1)])
    TN = len(df[(df[f'event{i}']==0)&(df[f'pred_y{i}']==0)])
    FN = len(df[(df[f'event{i}']==1)&(df[f'pred_y{i}']==0)])
    FP = len(df[(df[f'event{i}']==0)&(df[f'pred_y{i}']==1)])


    result_i = performance(TP=TP, TN=TN, FP=FP, FN=FN)
    
    ewma_result = pd.concat([ewma_result, result_i])
    
ewma_result.index = [f"result{i}" for i in range(1, 1001)]

ewma_result.head(2)
```

```python
mean_ewma = pd.DataFrame(index=['mean'])

mean_ewma['mean_sens'] = ewma_result['Sensitivity'].mean()
mean_ewma['mean_spec'] = ewma_result['Specificity'].mean()
mean_ewma['mean_accu'] = ewma_result['Accuracy'].mean()
mean_ewma['mean_ppv'] = ewma_result['PPV'].mean()
mean_ewma['mean_npv'] = ewma_result['NPV'].mean()

mean_ewma
```
