```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

#### 난수 생성 및 plot

```python
np.random.seed(5)

data_80 = np.random.poisson(lam=80, size = 10000)

print(data_80[:5]);print(min(data_80));print(max(data_80))

sns.displot(data_80, kind = 'kde')

plt.title('lambda 80')
```

```python
pd.Series(data_80).plot()
```

##### 웨어러블 기기에서 뽑히는 데이터 셋처럼 만들기

Time Frequency
- B: business day, ie., Mon. - Fri
- D: daily
- W: weekly
- M: monthly
- A: annual
- Q: quarterly


```python
# 일별
np.random.seed(5)

df_80_d = pd.DataFrame({'update_time': pd.date_range(start='01-01-2000',
                                      periods=5000,
                                      freq='D'),
                   'heart_rate': np.random.poisson(lam=80, size = 5000)})
df_80_d.describe()
```

```python
# 시간대별
np.random.seed(5)

df_80_h = pd.DataFrame({'update_time': pd.date_range(start='01-01-2000',
                                      periods=10000,
                                      freq='H'),
                   'heart_rate': np.random.poisson(lam=80, size = 10000)})
df_80_h.head()
```

```python
df_70_h = pd.DataFrame({'update_time': pd.date_range(start='01-01-2000',
                                      periods=10000,
                                      freq='H'),
                   'heart_rate': np.random.poisson(lam=70, size = 10000)})
df_70_h.head()
```

```python
df_60_h = pd.DataFrame({'update_time': pd.date_range(start='01-01-2000',
                                      periods=10000,
                                      freq='H'),
                   'heart_rate': np.random.poisson(lam=60, size = 10000)})
df_60_h.head()
```

```python
df_90_h = pd.DataFrame({'update_time': pd.date_range(start='01-01-2000',
                                      periods=10000,
                                      freq='H'),
                   'heart_rate': np.random.poisson(lam=90, size = 10000)})
df_90_h.head()
```

```python
df_60_h.to_csv('포아송_1만_단위h_60.csv', index=False)
df_70_h.to_csv('포아송_1만_단위h_70.csv', index=False)
df_80_h.to_csv('포아송_1만_단위h_80.csv', index=False)
df_90_h.to_csv('포아송_1만_단위h_90.csv', index=False)
```

```python
ax1 = plt.subplot(1, 1, 1)

plt.plot(df_90_h.update_time, df_90_h['heart_rate'], label='heart_rate')

plt.title('lambda 90 heart_rate')
plt.xlabel('Duration')
plt.ylabel('heart_rate')
plt.legend()
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
```

```python
ax1 = plt.subplot(1, 1, 1)

plt.plot(df_80_h.update_time, df_80_h['heart_rate'], label='heart_rate')

plt.title('lambda 80 heart_rate')
plt.xlabel('Duration')
plt.ylabel('heart_rate')
plt.legend()
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
```

```python
ax1 = plt.subplot(1, 1, 1)

plt.plot(df_70_h.update_time, df_70_h['heart_rate'], label='heart_rate')

plt.title('lambda 70 heart_rate')
plt.xlabel('Duration')
plt.ylabel('heart_rate')
plt.legend()
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
```

#### 구현한 ewma 적용하기

```python
from ewma import ewma

help(ewma)
'''
FUNCTIONS
    ewma(data, initial, gamma, k)
'''
```

```python
# lambda = 90
a1 = ewma(df_90_h, 95, 0.95, 3)
a2 = ewma(df_90_h, 80, 0.95, 3)
a3 = ewma(df_90_h, 85, 0.95, 3)
a4 = ewma(df_90_h, 75, 0.95, 3)
a5 = ewma(df_90_h, 65, 0.95, 3)
a6 = ewma(df_90_h, 95, 0.95, 2)
a7 = ewma(df_90_h, 80, 0.95, 2)
a8 = ewma(df_90_h, 85, 0.95, 2)
a9 = ewma(df_90_h, 75, 0.95, 2)
a10 = ewma(df_90_h, 65, 0.95, 2)


print(a1['outlier_yn'].unique())
print(a2['outlier_yn'].unique())
print(a3['outlier_yn'].unique())
print(a4['outlier_yn'].unique())
print(a5['outlier_yn'].unique())
print(a6['outlier_yn'].unique())
print(a7['outlier_yn'].unique())
print(a8['outlier_yn'].unique())
print(a9['outlier_yn'].unique())
print(a10['outlier_yn'].unique())
```
