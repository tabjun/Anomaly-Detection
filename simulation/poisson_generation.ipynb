```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(5)

data_80 = np.random.poisson(lam=80, size = 10000)

print(data_80[:5]);print(min(data_80));print(max(data_80))

sns.displot(data_80, kind = 'kde')

plt.title('lambda 80')
```

```python
np.random.seed(5)

data_70 = np.random.poisson(lam=70, size = 10000)

print(data_70[:5]);print(min(data_70));print(max(data_70))

sns.displot(data_70, kind = 'kde')

plt.title('lambda 70')
```

```python
data_60 = pd.DataFrame(data_60)
data_70 = pd.DataFrame(data_70)
data_80 = pd.DataFrame(data_80)
data_90 = pd.DataFrame(data_90)
```
