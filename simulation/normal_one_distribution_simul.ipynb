```python
# 임시 데이터 만들기
df = pd.DataFrame()

# 변수와 표본 개수 설정
l_size, r_size, h_size = 10, 950, 40    # 표본 개수 (low, 정상, high)
l_mean, r_mean, h_mean = 40, 80, 120    # 평균
std = 20                                # 표준편차

# 각 변수 생성과 데이터프레임 병합
for i in range(1, 1001):
   # 각 변수별로 다른 random seed 설정
   np.random.seed(i)
    
   # 정규분포 랜덤 추출
   l_data = np.random.normal(l_mean, std, l_size)  # low (비정상)
   l_data = l_data.astype(int)
   l_anomaly = np.ones(10)
    
   r_data = np.random.normal(r_mean, std, r_size)  # 정상
   r_data = r_data.astype(int)
   r_anomaly = np.zeros(950)
    
   h_data = np.random.normal(h_mean, std, h_size) # high (비정상)
   h_data = h_data.astype(int)
   h_anomaly = np.full(40, 2)

   # 데이터 합치기
   low = np.vstack((l_data, l_anomaly))
   raw = np.vstack((r_data, r_anomaly))
   high = np.vstack((h_data, h_anomaly))

   # 전체 데이터 합치기
   var = np.hstack((low, raw, high))

   # 변수명 생성 
   var_name = f'heart_rate{i}'
   anomaly_name = f'anomaly_yn{i}'
    
   # 데이터프레임에 변수 추가
   df[var_name] = var[0, :]
   df[anomaly_name] = var[1, :]

   # 변수(본포{i})마다 행 순서 바꾸기
   np.random.seed(i)
   df = df.sample(frac=1)
```

##### 일주일 + 3일(342) 만들기
- 약 50개 정도의 noise

```python
# 한 분포에서 추출하기
# train
df_tr = pd.DataFrame()

# 각 변수 생성과 데이터프레임 병합
for i in range(1, 1001):
   # 각 변수별로 다른 random seed 설정
   np.random.seed(i)
   
   # 정규분포 랜덤 추출
   # 일주일치 1008개
   r_data = np.random.normal(80, 10, 1008)  # 정상
   r_data = r_data.astype(int)    
   
   # 변수명 생성 
   var_name = f'heart_rate{i}'
   
   # 데이터프레임에 변수 추가
   df_tr[var_name] = r_data
   
   # event로 변수 생성
   df_tr[f'event{i}'] = df_tr[var_name].apply(lambda x: 1 if (x < 60 )|(x > 100) else 0 if 60 <= x <= 100 else x)
   
   # 변수(본포{i})마다 행 순서 바꾸기
   np.random.seed(i)
   df_tr = df_tr.sample(frac=1)
   
print(df_tr.shape)
```

```python
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

for i in range(1,1001):
    sns.kdeplot(df_tr[f'heart_rate{i}'])
    
plt.title('heart_rate 정규분포')
plt.show()
```
