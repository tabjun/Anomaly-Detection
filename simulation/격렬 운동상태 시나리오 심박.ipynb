```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
```

```python
# update_time 데이터프레임 생성, 10분 단위로 생성
date = pd.date_range('2023-10-01 00:00:00', periods=1350, freq='10T')

# 임시 데이터 만들기
q1 = pd.DataFrame()
q1['a'] = 1

# 데이터 합치기
q = pd.DataFrame(q1, index = date)

# datetimeindex를 변수로 변경
q.reset_index(drop=False, inplace=True)

# 필요없는 변수 삭제
del q['a']

# 변수명 변경
q.rename(columns = {'index' : 'update_time'}, inplace = True)
q
```

# 적당한 운동 데이터 만들기

### 청년 만들기
- 일주일 + 3일(342) 만들기
- 노이즈 5%, 10%

```python
178+80
258/2
```

```python
print(pd.DataFrame(np.random.normal(129, 24, 1350)).describe())
r_data = np.random.normal(129, 24, 1350)

r = pd.DataFrame()
r['a'] = r_data
print(len(r[(r['a']<80)|(r['a']>178)]))
```

```python
젊음 = pd.DataFrame()

# 각 변수 생성과 데이터프레임 병합
for i in range(1, 1001):
   # 각 변수별로 다른 random seed 설정
   np.random.seed(i)
   
   # 정규분포 랜덤 추출
   # 일주일치 1008개
   r_data = np.random.normal(129, 24, 1350) 
   r_data = r_data.astype(int)    
   
   # 변수명 생성 
   var_name = f'heart_rate{i}'
   
   # 데이터프레임에 변수 추가
   젊음[var_name] = r_data
   
   # event로 변수 생성
   젊음[f'event{i}'] = 젊음[var_name].apply(lambda x: 1 if (x < 80 )|(x > 178) else 0 if 80 <= x <= 178 else x)
   
   # 변수(본포{i})마다 행 순서 바꾸기
   np.random.seed(i)
   젊음 = 젊음.sample(frac=1)
   
print(젊음.shape)
```
